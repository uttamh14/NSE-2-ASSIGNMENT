```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(
    df['reviews_per_month'].median()
)
df
```

```python
df['name'] = df['name'].fillna('Unknown')
df
```

```python
df['neighbourhood_group'] = df['neighbourhood_group'].str.strip().str.title()
df['room_type'] = df['room_type'].str.strip().str.title()
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df
```

```python
df.isna().sum().sort_values(ascending=False)
df
```

```python
df['expensive_listing'] = df['price'] > 200
df
```

```python
df['review_level'] = pd.cut(
    df['number_of_reviews'],
    bins=[-1, 10, 50, 100, df['number_of_reviews'].max()],
    labels=['Low', 'Medium', 'High', 'Very High']
)
df
```

```python
df['last_review_year'] = df['last_review'].dt.year
df
```

```python
df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
df
```

```python
df['room_type'].value_counts()
df
```

```python
pd.pivot_table(
    df,
    values='price',
    index='neighbourhood_group',
    columns='room_type',
    aggfunc='mean'
)
df
```
